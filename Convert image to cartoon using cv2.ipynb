```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2 
import numpy as np 
```

```python
image = mpimg.imread('download.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
# change the color image to gray image
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray, cmap = 'gray')
```

```python
 
image_gray.shape
```

```python
# saving the image
cv2.imwrite('image_gray.jpg', image_gray)
```

```python
# sigma_s range(0, 200)
# sigma_r range(0, 1)
cartoon = cv2.stylization(image, sigma_s = 200, sigma_r = 0.3)
```

```python
plt.imshow(cartoon)
```

```python
edge = cv2.edgePreservingFilter(image, flags=1, sigma_s=60, sigma_r=0.4)
```

```python
plt.imshow(edge)
```

```python
import cv2
detail = cv2.detailEnhance(image, sigma_s=10, sigma_r=0.15)
```

```python
plt.imshow(detail)
```
